### 利差与汇率

```python
# 获取当前工作目录
cwd = os.getcwd()

# 指定要读取数据的目录
data_dir = os.path.join(cwd, 'data_xlsx')

data_file = '即期汇率_美元兑人民币.xlsx'
data_path = os.path.join(data_dir, data_file)
FXRate_table = pd.read_excel(data_path).dropna()
FXRate_table.columns = ['Time', 'US_CHN_rate']
FXRate_table['Time'] = pd.to_datetime(FXRate_table['Time'])

data_file = '美国_国债收益率_10年.xlsx'
data_path = os.path.join(data_dir, data_file)
US_Y10_table = pd.read_excel(data_path).dropna()
US_Y10_table.columns = ['Time', 'US_Y10']
US_Y10_table['Time'] = pd.to_datetime(US_Y10_table['Time'])

data_file = '美国_国债收益率_2年.xlsx'
data_path = os.path.join(data_dir, data_file)
US_Y2_table = pd.read_excel(data_path).dropna()
US_Y2_table.columns = ['Time', 'US_Y2']
US_Y2_table['Time'] = pd.to_datetime(US_Y2_table['Time'])

# 指定要读取数据的目录
data_dir = os.path.join(cwd, 'data_csv')

data_file = '中国十年期国债收益率历史数据.csv'
data_path = os.path.join(data_dir, data_file)
CHN_Y10_table = pd.read_csv(data_path)
CHN_Y10_table = CHN_Y10_table.iloc[:,0:2].dropna()
CHN_Y10_table.columns = ['Time', 'CHN_Y10']
CHN_Y10_table['Time'] = pd.to_datetime(CHN_Y10_table['Time'])

data_file = '中国二年期国债收益率历史数据.csv'
data_path = os.path.join(data_dir, data_file)
CHN_Y2_table = pd.read_csv(data_path)
CHN_Y2_table = CHN_Y2_table.iloc[:,0:2].dropna()
CHN_Y2_table.columns = ['Time', 'CHN_Y2']
CHN_Y2_table['Time'] = pd.to_datetime(CHN_Y2_table['Time'])

# 计算yearly_change_rate
FXRate_table['yearly_change_rate'] = (FXRate_table['US_CHN_rate'] - FXRate_table['US_CHN_rate'].shift(365)) / FXRate_table['US_CHN_rate'].shift(365)

# 删除缺失值
US_CHN_rate_table = FXRate_table.dropna()

# 合并数据框
Total2 = US_CHN_rate_table.merge(US_Y10_table, on='Time') \
    .merge(US_Y2_table, on='Time') \
    .merge(CHN_Y10_table, on='Time') \
    .merge(CHN_Y2_table, on='Time')

# 计算新变量
Total2['interest_spread_Y10'] = Total2['US_Y10'] - Total2['CHN_Y10']
Total2['interest_spread_Y2'] = Total2['US_Y2'] - Total2['CHN_Y2']
Total2['interest_spread_Y10_mod'] = Total2['US_Y10'] - Total2['CHN_Y10'] - Total2['yearly_change_rate']
Total2['interest_spread_Y2_mod'] = Total2['US_Y2'] - Total2['CHN_Y2'] - Total2['yearly_change_rate']

mean_interest_spread_Y10 = Total2["interest_spread_Y10"].mean()
sd_interest_spread_Y10 = Total2["interest_spread_Y10"].std()

mean_interest_spread_Y2 = Total2["interest_spread_Y2"].mean()
sd_interest_spread_Y2 = Total2["interest_spread_Y2"].std()

mean_interest_spread_Y10_mod = Total2["interest_spread_Y10_mod"].mean()
sd_interest_spread_Y10_mod = Total2["interest_spread_Y10_mod"].std()

mean_interest_spread_Y2_mod = Total2["interest_spread_Y2_mod"].mean()
sd_interest_spread_Y2_mod = Total2["interest_spread_Y2_mod"].std()

mean_US_CHN_rate = Total2["US_CHN_rate"].mean()
sd_US_CHN_rate = Total2["US_CHN_rate"].std()

# 计算US_CHN_rate_z
Total2['US_CHN_rate_z'] = (Total2['US_CHN_rate'] - Total2['US_CHN_rate'].mean()) / Total2['US_CHN_rate'].std()
```

```python
# 计算y2
y2 = Total2['US_CHN_rate_z'] * Total2['interest_spread_Y10_mod'].std() + Total2['interest_spread_Y10_mod'].mean()

# 绘制折线图
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(Total2['Time'], Total2['interest_spread_Y10_mod'], label='美国中国10年期国债收益率之差%（经过汇率调整）', color='#007896FF', linewidth=1)
ax.plot(Total2['Time'], y2, label='美元/人民币', color='#db6968', linewidth=1)

# 设置x轴刻度
ax.xaxis.set_major_locator(MonthLocator(interval=3))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m-%d'))
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# 设置y轴次坐标轴
def y2y2(y):
    return (y - Total2['interest_spread_Y10_mod'].mean()) / Total2['interest_spread_Y10_mod'].std() * Total2['US_CHN_rate'].std() + Total2['US_CHN_rate'].mean()
ax2 = ax.secondary_yaxis('right', functions=(y2y2, y2y2))

# 设置图例和坐标轴标签
ax.legend(loc='upper center')
ax.set_xlabel('')
ax.set_ylabel('10年期国债收益率之差%（汇率调整）')
ax2.set_ylabel('汇率')

# 设置标题
ax.set_title('美国中国10年期国债收益率之差与美元兑人民币汇率')

# 显示图形
plt.show()
```

```python
# 计算y2
y2 = Total2['US_CHN_rate_z'] * Total2['interest_spread_Y2_mod'].std() + Total2['interest_spread_Y2_mod'].mean()

# 绘制折线图
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(Total2['Time'], Total2['interest_spread_Y2_mod'], label='美国中国2年期国债收益率之差%（经过汇率调整）', color='#007896FF', linewidth=1)
ax.plot(Total2['Time'], y2, label='美元/人民币', color='#db6968', linewidth=1)

# 设置x轴刻度
ax.xaxis.set_major_locator(MonthLocator(interval=3))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m-%d'))
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# 设置y轴次坐标轴
def y2y2(y):
    return (y - Total2['interest_spread_Y2_mod'].mean()) / Total2['interest_spread_Y2_mod'].std() * Total2['US_CHN_rate'].std() + Total2['US_CHN_rate'].mean()
ax2 = ax.secondary_yaxis('right', functions=(y2y2, y2y2))

# 设置图例和坐标轴标签
ax.legend(loc='upper center')
ax.set_xlabel('')
ax.set_ylabel('2年期国债收益率之差%（汇率调整）')
ax2.set_ylabel('汇率')

# 设置标题
ax.set_title('美国中国2年期国债收益率之差与美元兑人民币汇率')

# 显示图形
plt.show()
```

```python
FXRate_table = FXRate_table[FXRate_table['Time'].dt.year >= split_year]
US_Y10_table = US_Y10_table[US_Y10_table['Time'].dt.year >= split_year]
US_Y2_table = US_Y2_table[US_Y2_table['Time'].dt.year >= split_year]
CHN_Y10_table = CHN_Y10_table[CHN_Y10_table['Time'].dt.year >= split_year]
CHN_Y2_table = CHN_Y2_table[CHN_Y2_table['Time'].dt.year >= split_year]

# 计算yearly_change_rate
FXRate_table['yearly_change_rate'] = (FXRate_table['US_CHN_rate'] - FXRate_table['US_CHN_rate'].shift(365)) / FXRate_table['US_CHN_rate'].shift(365)

# 删除缺失值
US_CHN_rate_table = FXRate_table.dropna()

# 合并数据框
Total2 = US_CHN_rate_table.merge(US_Y10_table, on='Time') \
    .merge(US_Y2_table, on='Time') \
    .merge(CHN_Y10_table, on='Time') \
    .merge(CHN_Y2_table, on='Time')

# 计算新变量
Total2['interest_spread_Y10'] = Total2['US_Y10'] - Total2['CHN_Y10']
Total2['interest_spread_Y2'] = Total2['US_Y2'] - Total2['CHN_Y2']
Total2['interest_spread_Y10_mod'] = Total2['US_Y10'] - Total2['CHN_Y10'] - Total2['yearly_change_rate']
Total2['interest_spread_Y2_mod'] = Total2['US_Y2'] - Total2['CHN_Y2'] - Total2['yearly_change_rate']

mean_interest_spread_Y10 = Total2["interest_spread_Y10"].mean()
sd_interest_spread_Y10 = Total2["interest_spread_Y10"].std()

mean_interest_spread_Y2 = Total2["interest_spread_Y2"].mean()
sd_interest_spread_Y2 = Total2["interest_spread_Y2"].std()

mean_interest_spread_Y10_mod = Total2["interest_spread_Y10_mod"].mean()
sd_interest_spread_Y10_mod = Total2["interest_spread_Y10_mod"].std()

mean_interest_spread_Y2_mod = Total2["interest_spread_Y2_mod"].mean()
sd_interest_spread_Y2_mod = Total2["interest_spread_Y2_mod"].std()

mean_US_CHN_rate = Total2["US_CHN_rate"].mean()
sd_US_CHN_rate = Total2["US_CHN_rate"].std()

# 计算US_CHN_rate_z
Total2['US_CHN_rate_z'] = (Total2['US_CHN_rate'] - Total2['US_CHN_rate'].mean()) / Total2['US_CHN_rate'].std()
```
