```python
import yfinance as yf

def get_data_yfinance(start_date, end_date, symbol): # symbol是指标在Yahoo Finance上的代码
    ticker = yf.Ticker(symbol)
    ticker_data = ticker.history(start=start_date, end=end_date)
    ticker_data.reset_index(inplace=True)
    ticker_data['Date'] = ticker_data['Date'].dt.date
    return ticker_data

SPX_table = get_data_yfinance("2015-01-01", today, '^GSPC') # 标准普尔500指数
SPX_table.to_csv("data_csv/美国_标准普尔500指数.csv", index = None, header = True)

USDX_table = get_data_yfinance("2015-01-01", today, "DX-Y.NYB")
USDX_table.to_csv("data_csv/美元指数.csv", index = None, header = True)

USDCNY_table = get_data_yfinance("2015-01-01", today, "USDCNY=X")
USDCNY_table.to_csv("data_csv/USDCNY.csv", index = None, header = True)

USDHKD_table = get_data_yfinance("2007-01-01", today, "HKD=X")
USDHKD_table.to_csv("data_csv/USDHKD.csv", index = None, header = True)
```

```python
# 获取当前工作目录
cwd = os.getcwd()

# 判断是否存在figure文件夹，如果不存在就创建该文件夹
if not os.path.exists('figure'):
    os.makedirs('figure')

# 指定要读取数据的目录
data_dir = os.path.join(cwd, 'data_xlsx')

#data_file = '即期汇率_美元兑人民币.xlsx'
#data_path = os.path.join(data_dir, data_file)
#ER = pd.read_excel(data_path).dropna()
ER = pd.read_csv("data_csv/USDCNY.csv")
ER = ER.loc[:,['Date', 'Close']].dropna()
ER.columns = ['Time', 'exRate']
ER['Time'] = pd.to_datetime(ER['Time'])

#data_file = '美元指数.xlsx'
#data_path = os.path.join(data_dir, data_file)
#USDX = pd.read_excel(data_path).dropna()
USDX = pd.read_csv("data_csv/美元指数.csv")
USDX = USDX.loc[:,['Date', 'Close']].dropna()
USDX.columns = ['Time', 'Price_USDX']
USDX['Time'] = pd.to_datetime(USDX['Time'])

# 指定要读取数据的目录
# data_dir = os.path.join(cwd, 'data_csv')

# data_file = '000300.SH-行情统计.csv'
# data_path = os.path.join(data_dir, data_file)
# SH = pd.read_csv(data_path)
# SH = SH.loc[:,['交易日期', '收盘价']].dropna()
# SH.columns = ['Time', 'Price_SH']
# SH['Time'] = pd.to_datetime(SH['Time'])
# SH['Price_SH'] = SH['Price_SH'].str.replace(',', '').astype(float)

#data_file = '沪深300指数.xlsx'
#data_path = os.path.join(data_dir, data_file)
#SH = pd.read_excel(data_path).dropna()
SH = pd.read_csv("data_csv/沪深300指数.csv")
SH = SH.loc[:,['日期', '收盘']].dropna()
SH.columns = ['Time', 'Price_SH']
SH['Time'] = pd.to_datetime(SH['Time'])

# data_file = 'SPX.GI-行情统计.csv'
# data_path = os.path.join(data_dir, data_file)
# SPX = pd.read_csv(data_path)
# SPX = SPX.loc[:,['交易日期', '收盘价']].dropna()
# SPX.columns = ['Time', 'Price_SPX']
# SPX['Time'] = pd.to_datetime(SPX['Time'])
# SPX['Price_SPX'] = SPX['Price_SPX'].str.replace(',', '').astype(float)

#data_file = '美国_标准普尔500指数.xlsx'
#data_path = os.path.join(data_dir, data_file)
#SPX = pd.read_excel(data_path).dropna()
SPX = pd.read_csv("data_csv/美国_标准普尔500指数.csv")
SPX = SPX.loc[:,['Date', 'Close']].dropna()
SPX.columns = ['Time', 'Price_SPX']
SPX['Time'] = pd.to_datetime(SPX['Time'])

#data_file = 'MSCI新兴市场.xlsx'
#data_path = os.path.join(data_dir, data_file)
#MSCI = pd.read_excel(data_path).dropna()
MSCI = pd.read_csv('data_csv/MSCI新兴市场.csv').dropna()
MSCI.columns = ['Time', 'Price_MSCI']
MSCI['Time'] = pd.to_datetime(MSCI['Time'])

Total1 = ER.merge(SH, on='Time') \
    .merge(SPX, on='Time') \
    .merge(USDX, on='Time') \
    .merge(MSCI, on ='Time')

# Calculate stockRate
Total1['stockRate'] = Total1['Price_SPX'] / Total1['Price_SH'] * Total1['exRate']

"""
# Calculate means and standard deviations
mean_exRate = Total1['exRate'].mean()
sd_exRate = Total1['exRate'].std()

mean_stockRate = Total1['stockRate'].mean()
sd_stockRate = Total1['stockRate'].std()

mean_Price_USDX = Total1['Price_USDX'].mean()
sd_Price_USDX = Total1['Price_USDX'].std()

mean_Price_SH = Total1['Price_SH'].mean()
sd_Price_SH = Total1['Price_SH'].std()
"""

# 计算exRate_z
Total1['exRate_z'] = (Total1['exRate'] - Total1['exRate'].mean()) / Total1['exRate'].std()

# 计算USDX_z
Total1['USDX_z'] = (Total1['Price_USDX'] - Total1['Price_USDX'].mean()) / Total1['Price_USDX'].std()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib.dates import DateFormatter, MonthLocator

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] =False

# 计算y2
y2 = Total1['exRate_z'] * Total1['stockRate'].std() + Total1['stockRate'].mean()

# 绘制折线图
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(Total1['Time'], Total1['stockRate'], label='美国标普500/沪深300（以美元计价）', color='#007896FF', linewidth=1)
ax.plot(Total1['Time'], y2, label='美元/人民币', color='#db6968', linewidth=1)

# 设置x轴刻度
ax.xaxis.set_major_locator(MonthLocator(interval=3))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m-%d'))
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# 设置y轴次坐标轴
def y2y2(y):
    return (y - Total1['stockRate'].mean()) / Total1['stockRate'].std() * Total1['exRate'].std() + Total1['exRate'].mean()
ax2 = ax.secondary_yaxis('right', functions=(y2y2, y2y2))

# 设置图例和坐标轴标签
ax.legend(loc='upper center')
ax.set_xlabel('')
ax.set_ylabel('美国标普500/沪深300（以美元计价）')
ax2.set_ylabel('汇率')

# 设置标题
ax.set_title('标普500和沪深300比值与美元兑人民币汇率')

# 保存图片
plt.savefig("figure/标普500和沪深300比值与美元兑人民币汇率（长线）.png")

# 显示图形
plt.show()
```
